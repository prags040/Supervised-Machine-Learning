```python
# Importing all libraries required in this notebook
import pandas as pd
import numpy as np  
import matplotlib.pyplot as plt  
%matplotlib inline
```

## Importing Data from te given Dataset URL

```python
# Reading data from remote link
url = "http://bit.ly/w-data"
s_data = pd.read_csv(url)
print("Data imported successfully")

s_data.head(10)
```

## 2-D Plotting Of Scores

```python
# 2-D Plotting the distribution of scores
s_data.plot(x='Hours', y='Scores', style='o')  
plt.title('Hours vs Percentage')  
plt.xlabel('Hours Studied')  
plt.ylabel('Percentage Score')  
plt.show()
```

## Dividing Data into 'attributes' & 'Labels'

```python
# dividing the data into "attributes" (inputs) and "labels" (outputs).
X = s_data.iloc[:, :-1].values  
y = s_data.iloc[:, 1].values
```

## Splitting Data into Training and Test sets

```python
# split this data into training and test sets.
from sklearn.model_selection import train_test_split  
X_train, X_test, y_train, y_test = train_test_split(X, y, 
                            test_size=0.2, random_state=0) 
```

## Training Algorithm

```python
#Training Algo
from sklearn.linear_model import LinearRegression  
regressor = LinearRegression()  
regressor.fit(X_train, y_train) 

print("Training completed successfully.")
```

## Plotting Of Regression Line 

```python
# Plotting the regression line
line = regressor.coef_*X+regressor.intercept_

# Plotting for the test data
plt.scatter(X, y)
plt.plot(X, line);
plt.show()
```

## Prediction

```python
#PREDICTION
# Testing data - In Hours
print(X_test) 
# Predicting the scores
y_pred = regressor.predict(X_test) 
```

```python
# Comparing Actual vs Predicted
df = pd.DataFrame({'Actual': y_test, 'Predicted': y_pred})  
df 
```

## Testing User Input  

```python
#Lets test it with our own data
hours = 9.25
print("No of Hours student studied : ", hours)
print("Predicted Score of student depending upon the hours studied = ", regressor.predict(np.array(hours).reshape(1,-1))[0])
```

```python
#Evaluation step
from sklearn import metrics  
print('Mean Absolute Error:', 
      metrics.mean_absolute_error(y_test, y_pred)) 
```
